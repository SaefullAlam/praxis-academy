# Membuat mesin ATM

Pada kasus kali ini adalah membuat program ATM sederhana

```python
userdata = [
    {
        'id'       : '1111',
        'rekening' : '123456789',
        'username' : 'Tono',
        'pin'      : '12345',
        'saldo'    : 100000
    },
    {
        'id'       : '1112',
        'rekening' : '145689023',
        'username' : 'Boni',
        'pin'      : '55555',
        'saldo'    : 50000
    },
    {
        'id'       : '1113',
        'rekening' : '527436982',
        'username' : 'Siti',
        'pin'      : '33333',
        'saldo'    : 0
    },
]
```

```python
class ATM:
    def __init__(self,database):
        self.userdata    = database 
        self.run         = True
        self.session     = False
        self.position      = None
        self.session_user  = None
        
        while self.run:
            self.main()
    
    def login(self):
        while not self.session:
            temp_id  = input('Masukkan id : ')
            temp_pin = input('Masukkan pin : ')
            for user in self.userdata:
                if user['id']==temp_id and user['pin']==temp_pin:
                    print('Login sukses')
                    setattr(self,'session',True)
                    setattr(self,'session_user',user)
                    setattr(self,'position','dashboard')
                    break
            if self.session:
                break
            print('id/password salah')
    
    def dashboard(self):
        print('''
                 1. Cek saldo
                 2. Isi saldo
                 3. Transfer
                 4. Isi pulsa
                 5. Top up E-Wallet
                ''')
        pilihan = ['Cek saldo','Isi saldo','Transfer','Isi pulsa','Top up E-Wallet','Pembayaran'] 
        pilih   = int(input('Pilih Menu (1-5) : '))-1
        setattr(self,'position',pilihan[pilih])
    
    def cek_saldo(self):
        saldo = self.session_user['saldo']
        print(f'Saldo anda adalah: {saldo}')
        setattr(self,'position','dashboard')
        self.exit()
    
    def isi_saldo(self):
        tambah = int(input("Masukkan jumlah top up saldo: "))
        self.session_user['saldo']+=tambah
        saldo = self.session_user['saldo']
        print(f'Saldo anda adalah: {saldo}')
        self.exit()
    
    def transfer(self):
        rek_tujuan  = input("Masukkan rekening tujuan :")
        jumlah      = int(input("Masukkan nominal transfer : "))
        user_tujuan = None
        
        for user in self.userdata:
            if user['rekening']==rek_tujuan:
                user_tujuan = user
        if self.session_user['saldo']<jumlah:
                print('saldo anda tidak mencukupi')
        else:
            try:
                user_tujuan['saldo']+=jumlah
                self.session_user['saldo']-=jumlah
                print(user_tujuan['username'])
                print(user_tujuan['saldo'])
            except:
                print('Rekening tidak valid')
                
        self.exit()
    
    def isi_pulsa(self):
        provider = input('Pilih provider : ') 
        nomor    = input('Masukkan nomor : ')
        nominal  = int(input('Masukkan nominal :'))
        if self.session_user['saldo']<nominal:
                print('saldo anda tidak mencukupi')
        else:
            try:
                self.session_user['saldo']-=nominal
                print(f'''
                        Nomor    : {nomor}
                        Provider : {provider}
                        Nominal  : {nominal}''')
            except:
                print('Nomor tidak valid')
        self.exit()
    
    def top_up_ewallet(self):
        ewallet  = input('Pilih E-Wallet : ') 
        nomor    = input('Masukkan nomor : ')
        nominal  = int(input('Masukkan nominal :'))
        if self.session_user['saldo']<nominal:
                print('saldo anda tidak mencukupi')
        else:
            try:
                self.session_user['saldo']-=nominal
                print(f'''
                        Nomor     : {nomor}
                        E-Wallet  : {ewallet}
                        Nominal   : {nominal}''')
            except:
                print('Nomor tidak valid')
        self.exit()
                
    
    def exit(self):
        print('''Apakah anda ingin melakukan transaksi lagi?
                          1. Ya
                          2. Tidak ''')
        keluar = input()
        if keluar=='1':
            setattr(self,'position','dashboard')
            setattr(self,'run',True)
        else :
            print('Terimakasih telah menggunakan jasa kami')
            setattr(self,'session',False)
            setattr(self,'session_user',None)
            setattr(self,'run',False)
    
    def main(self):
        print('Selamat datang di Mesin ATM')
        self.login()
        
        while self.session:
            if self.position=='dashboard':
                self.dashboard()
            elif self.position=='Cek saldo':
                self.cek_saldo()
            elif self.position=='Isi saldo':
                self.isi_saldo()
            elif self.position=='Transfer':
                self.transfer()
            elif self.position=='Isi pulsa':
                self.isi_pulsa()
            elif self.position=='Top up E-Wallet':
                self.top_up_ewallet()
            else:
                print('gagal')
            
    
```

```python
mesin1 = ATM(userdata)
```
