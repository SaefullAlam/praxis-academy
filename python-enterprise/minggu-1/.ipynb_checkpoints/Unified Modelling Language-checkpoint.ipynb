UML (Unified Modelling Language) adalah bahasa standar penulisan blueprint sebuah perangkat lunak yang meliputi visualisasi, perancangan, hingga dokumentasi sistem. Penerapan UML pada sistem berorientasi objek yaitu sebagai metode perancangan model secara visual. sebuah perangkat lunak memiliki banyak komponen maupun faktor penyusun seperti scalability, robustness, security, dan lain lain. Banyaknya faktor yang ada berpotensi terjadi kesalahpahaman. UML berguna untuk memudahkan proses pengembangan perangkat lunak serta mencegah adanya kesalahpahaman antar pengembang ataupun antara pengembang dan klien karena menyediakan gambaran dan alir yang jelas secara visual.

1. Dapat memberikan bahasa pemodelan visual atau gambar kepada para pengguna dari berbagai macam pemrograman maupun proses umum rekayasa.
2. Menyatukan informasi-informasi terbaik yang ada dalam pemodelan.
3. Memberikan suatu gambaran model atau sebagai bahasa pemodelan visual yang ekspresif dalam pengembangan sistem.
4. Tidak hanya menggambarkan model sistem software saja, namun dapat memodelkan sistem berorientasi objek.
5. Mempermudah pengguna untuk membaca suatu sistem.
6. Berguna sebagai blueprint, jelas ini nantinya menjelaskan informasi yang lebih detail dalam perancangan berupa coding suatu program.
